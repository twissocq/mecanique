```python
import numpy as np
import matplotlib.pyplot as plt

def calculer_trajectoire(vitesse_initiale, angle, temps_total, pas_de_temps):
    # Convertir l'angle de degrés en radians
    angle_rad = np.radians(angle)

    # Calculer les composantes de la vitesse initiale
    vitesse_x = vitesse_initiale * np.cos(angle_rad)
    vitesse_y = vitesse_initiale * np.sin(angle_rad)

    # Initialiser les listes pour stocker les positions
    temps = np.arange(0, temps_total, pas_de_temps)
    x = []
    y = []

    # Calculer la position à chaque instant
    for t in temps:
        pos_x = vitesse_x * t
        pos_y = vitesse_y * t - 0.5 * 9.81 * t**2  # 9.81 m/s² est l'accélération due à la gravité
        x.append(pos_x)
        y.append(pos_y)
        if pos_y < 0:  # Arrêter si la bille touche le sol
            break
    # Convertir les listes en tableaux numpy    
    x = np.array(x)
    y = np.array(y)
    # S'assurer que les tableaux ont la même longueur
    min_length = min(len(x), len(y))
    x = x[:min_length]
    y = y[:min_length]
    # Retourner les positions       
    return x, y

# Conditions initiales
vitesse_initiale = 20  # en mètres par seconde
angle = 45  # en degrés
temps_total = 4  # en secondes
pas_de_temps = 0.01  # en secondes

# Calculer la trajectoire
x, y = calculer_trajectoire(vitesse_initiale, angle, temps_total, pas_de_temps)

# Tracer la trajectoire
plt.plot(x, y)
plt.title("Trajectoire d'un lancer de bille")
plt.xlabel("Distance horizontale (m)")
plt.ylabel("Hauteur (m)")
plt.grid(True)
plt.show()
```
